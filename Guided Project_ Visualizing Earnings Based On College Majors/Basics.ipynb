```python
import pandas as pd 
import matplotlib.pyplot as plt
from numpy import arange
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
#print(recent_grads.iloc[0])
#print(recent_grads.head(1))
#print(recent_grads.tail(1))
#print(recent_grads.describe())

recent_grads = recent_grads.dropna()
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter',
                  title='Median vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter',
                  title='Unemployment_rate vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter',
                  title='Full_time vs. Median', figsize=(5,10))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter',
                  title='ShareWomen vs. Unemployment_rate', figsize=(5,10))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter',
                  title='Men  vs. Median', figsize=(5,10))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter',
                  title='Women   vs. Median', figsize=(5,10))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen",
        "Unemployment_rate", "Men", "Women"]
fig = plt.figure(figsize = (5,32))
for r in range(len(cols)-1):
    ax = fig.add_subplot(len(cols), 1, r+1)
    ax = recent_grads[cols[r]].plot(kind = 'hist', rot = 40)
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))

scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(15,15))
```
